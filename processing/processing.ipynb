```python
import pandas as pd
from faker import Faker
from random import randint
from datetime import datetime, date, time
import math
from tqdm import tqdm
```

```python
# Setup faker to generate fake data
fake = Faker()
```

```python
# Read in the dataset
df = pd.read_csv('../data/Crime_Data_from_2020_to_Present.csv')
```

```python
# Clean the 'LOCATION' column
df['LOCATION'] = df.apply(lambda row: " ".join(row['LOCATION'].split()), axis=1)
```

```python
# Clean the 'Cross Street'  column
def clean_street(row):
#     print(row)
#     print(type(row))
    if type(row['Cross Street']) == str:
        return " ".join(row['Cross Street'].split())
    return row['Cross Street']
df['Cross Street'] = df.apply(lambda row: clean_street(row), axis=1)
```

```python
# Create INSERT statements for Users table
print('INSERT INTO Users(userID, name, age, sex, descent, email, password) VALUES')
for index in range(1000):
    user_id = fake.uuid4()
    name = fake.name()
    age = randint(0, 100)
    sex = 'M' if randint(0, 1) == 0 else 'F'
    descent = 'X'
    email = fake.email()
    password = fake.password()
    user_string = f"('{user_id}', '{name}', {age}, '{sex}', '{descent}', '{email}', '{password}'),"
    print(user_string)
```

```python
# Create INSERT statements for CrimeCodes table
print('INSERT INTO CrimeCodes(crimeCode, description) VALUES')
for crimeCode in df['Crm Cd'].unique():
    description = df[df['Crm Cd'] == crimeCode].iloc[0]['Crm Cd Desc']
    print(f'({crimeCode}, "{description}"),')
```

```python
# Create INSERT statements for Weapons table
print('INSERT INTO Weapons(weaponID, description) VALUES')
# print(df['Weapon Used Cd'][df['Weapon Used Cd'].notna()].astype(int).unique())
for weaponCode in sorted(df['Weapon Used Cd'][df['Weapon Used Cd'].notna()].astype(int).unique()):
    description = df[df['Weapon Used Cd'] == weaponCode].iloc[0]['Weapon Desc']
    print(f'({weaponCode}, "{description}"),')
```

```python
# Create INSERT statements for Victims table
print('INSERT INTO Victims(victimID, age, sex, descent) VALUES')
for index, row in df.iterrows():
    victim_id = index
    vict_age = row['Vict Age'] if row['Vict Age'] not in [0, -1] else 'NULL'
    vict_sex = row['Vict Sex'] if type(row['Vict Sex']) == str else 'X' 
    vict_descent = row['Vict Descent'] if type(row['Vict Descent']) == str else 'X' 
    victim_string = f"({victim_id}, {vict_age}, '{vict_sex}', '{vict_descent}'),"
    print(victim_string)
    if index == 1000:
        break
```

```python
# Create INSERT statements for Locations table
print('INSERT INTO Locations(locationID, latitude, longitude, area, street) VALUES')
for index, row in df.iterrows():
    location_id = index
    location = row['LOCATION']
    area = row['AREA']
    area_name = row['AREA NAME']
    street = row['Cross Street'] if type(row['Cross Street']) == str else 'NULL'
    latitude = row['LAT']
    longitude = row['LON']
    if street == 'NULL':
        location_string = f"({location_id}, {latitude}, {longitude}, '{area_name}', {street}),"
    else:
        location_string = f"({location_id}, {latitude}, {longitude}, '{area_name}', '{street}'),"
    print(location_string)
    if index == 1000:
        break
```
